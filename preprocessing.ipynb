Feature Selection:-

Now, we will do Feature Selection on the basis of some concepts like:
1. Zero Variance --> The feature which no variations like it have same values, it will be drop.
2. High Correlation --> If the two features are highly correlated and the impact on the output is same so we don't need both features, we can use only one.
3. Low Mutual Information --> If the shared information between feature and target is so low, we don't much need that feature. (MI < 0.01)
4. Low Permutation Importance --> If the feature has low permutation importance like their model accuracy reduce too much after shuffling their data, we don't need that feature.
(Imp < 0.0002)


```python
## Zero Variance
var = X.var()
zero_var = var[var == 0].index.tolist()
zero_var 
```

```python
## High Correlation
corr = X.corr().abs()
upper = corr.where(np.triu(np.ones(corr.shape), k=1).astype(bool))
high_corr_pairs = [(f1, f2) for f1 in upper.columns for f2 in upper.index if upper.loc[f2, f1] > 0.9]
high_corr_pairs
```

```python
upper
```

```python
## Low Mutual Information
from sklearn.feature_selection import mutual_info_classif

mi = mutual_info_classif(X, y, random_state=0)
mi_series = pd.Series(mi, index=X.columns)
low_mi = mi_series[mi_series < 0.01].index.tolist()
low_mi
```

```python
## Low Permutation Importance
from sklearn.inspection import permutation_importance
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(random_state=0)
rf.fit(X, y)
perm = permutation_importance(rf, X, y, n_repeats=10, random_state=0)
rf_imp = pd.Series(rf.feature_importances_, index=X.columns)
perm_series = pd.Series(perm.importances_mean, index=X.columns)
low_perm = perm_series[perm_series < 0.0002].index.tolist()

low_perm
```

```python
# From each high_corr pair, pick the lower RF importance feature to drop
drop_corr = []
for f1, f2, in high_corr_pairs:
    if rf_imp[f1] < rf_imp[f2]:
        drop_corr.append(f1)
    else:
        drop_corr.append(f2)
drop_corr = list(set(drop_corr))  # Remove duplicates

drop_corr
```

```python
# Combine all
features_to_drop = list(set(zero_var + low_mi + low_perm + drop_corr))
```

```python
features_to_drop
```

```python
training_data=training_data.drop(columns=features_to_drop,axis=1)
training_data.info()
```
